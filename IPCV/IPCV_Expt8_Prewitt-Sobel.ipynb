```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("4dir_lines.png", cv2.IMREAD_GRAYSCALE)
# cv2.imshow("Displayed Image",image)
```

#### Prewitt Operation

```python
prewitt_fx = np.zeros(image.shape[:2], dtype="uint8")
prewitt_fy = np.zeros(image.shape[:2], dtype="uint8")
prewitt_res = np.zeros(image.shape[:2], dtype="uint8")
```

```python
prewitt_fx_mask = np.array([[-1,-1,-1],[0,0,0],[1,1,1]])
prewitt_fy_mask = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
print("Prewitt Fx Mask: \n",prewitt_fx_mask)
print("Prewitt Fy Mask: \n",prewitt_fy_mask)
```

```python
for i in range(0, image.shape[0]-1):
  for j in range(0, image.shape[1]-1):
     prewitt_fx[i,j] = image[i-1,j-1]*prewitt_fx_mask[0,0] + image[i-1,j]*prewitt_fx_mask[0,1] + image[i-1,j+1]*prewitt_fx_mask[0,2] + image[i,j-1]*prewitt_fx_mask[1,0] + image[i,j]*prewitt_fx_mask[1,1] + image[i,j+1]*prewitt_fx_mask[1,2] + image[i+1,j-1]*prewitt_fx_mask[2,0] + image[i+1,j]*prewitt_fx_mask[2,1] + image[i+1,j+1]*prewitt_fx_mask[2,2]
     prewitt_fy[i,j] = image[i-1,j-1]*prewitt_fy_mask[0,0] + image[i-1,j]*prewitt_fy_mask[0,1] + image[i-1,j+1]*prewitt_fy_mask[0,2] + image[i,j-1]*prewitt_fy_mask[1,0] + image[i,j]*prewitt_fy_mask[1,1] + image[i,j+1]*prewitt_fy_mask[1,2] + image[i+1,j-1]*prewitt_fy_mask[2,0] + image[i+1,j]*prewitt_fy_mask[2,1] + image[i+1,j+1]*prewitt_fy_mask[2,2]

cv2_imshow(prewitt_fx)
print()
cv2_imshow(prewitt_fy)
```

```python
prewitt_res = prewitt_fx + prewitt_fy
cv2_imshow(prewitt_res)
```

```python
prewitt_res_mask = prewitt_fx_mask + prewitt_fy_mask
print(prewitt_res_mask)

prewitt_res_with_res_mask = np.zeros(image.shape[:2], dtype="uint8")
for i in range(0, image.shape[0]-1):
  for j in range(0, image.shape[1]-1):
     prewitt_res_with_res_mask[i,j] = image[i-1,j-1]*prewitt_res_mask[0,0] + image[i-1,j]*prewitt_res_mask[0,1] + image[i-1,j+1]*prewitt_res_mask[0,2] + image[i,j-1]*prewitt_res_mask[1,0] + image[i,j]*prewitt_res_mask[1,1] + image[i,j+1]*prewitt_res_mask[1,2] + image[i+1,j-1]*prewitt_res_mask[2,0] + image[i+1,j]*prewitt_res_mask[2,1] + image[i+1,j+1]*prewitt_res_mask[2,2]

cv2_imshow(prewitt_res_with_res_mask)
```
