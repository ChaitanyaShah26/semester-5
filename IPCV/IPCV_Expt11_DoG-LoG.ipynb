```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("pandora.png", cv2.IMREAD_GRAYSCALE)
# cv2.imshow("Displayed Image",image)
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

```python
print("Size of image: ",image.shape)
print("Datatype of image: ",image.dtype)
print("Channels of image: ",image.ndim)
print("Total number of pixels (Size of image):",image.size)
```

### Method-1

```python
sigma1 = 1
sigma2 = 0.4
gaussian_sigma1 = cv2.GaussianBlur(image, (3, 3), sigma1)
gaussian_sigma2 = cv2.GaussianBlur(image, (3, 3), sigma2)
```

```python
print("Sigma1 = 1:\n")
cv2_imshow(gaussian_sigma1)
```

```python
print("Sigma2 = 0.4:\n")
cv2_imshow(gaussian_sigma2)
```

```python
dog_image_m1 = gaussian_sigma1 - gaussian_sigma2
print("DoG Image:\n")
cv2_imshow(dog_image_m1)
```

### Method-2

```python
kernel1 = (5,5)
kernel2 = (7,7)
gaussian_kernel1 = cv2.GaussianBlur(image, kernel1, 1)
gaussian_kernel2 = cv2.GaussianBlur(image, kernel2, 1)
```

```python
print("Kernel1 = (5,5):\n")
cv2_imshow(gaussian_kernel1)
```

```python
print("Kernel2 = (7,7):\n")
cv2_imshow(gaussian_kernel2)
```

```python
dog_image_m2 = gaussian_kernel1 - gaussian_kernel2
print("DoG Image:\n")
cv2_imshow(dog_image_m2)
```
