```python
import numpy as np
import cv2

image = cv2.imread("hogwarts.jpg", cv2.IMREAD_GRAYSCALE)
# cv2.imshow("Displayed Image",image)
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

```python
print("Size of image: ",image.shape)
print("Datatype of image: ",image.dtype)
print("Channels of image: ",image.ndim)
print("Total number of pixels (Size of image):",image.size)
```

### High Pass Filter

```python
hpf_mask = np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
hpf_mask = hpf_mask/9
print(hpf_mask)
```

```python
hpf_img = np.zeros(image.shape[:2], dtype="uint8")
for i in range(0, image.shape[0]-1):
  for j in range(0, image.shape[1]-1):
     hpf_img[i,j] = image[i-1,j-1]*hpf_mask[0,0] + image[i-1,j]*hpf_mask[0,1] + image[i-1,j+1]*hpf_mask[0,2] + image[i,j-1]*hpf_mask[1,0] + image[i,j]*hpf_mask[1,1] + image[i,j+1]*hpf_mask[1,2] + image[i+1,j-1]*hpf_mask[2,0] + image[i+1,j]*hpf_mask[2,1] + image[i+1,j+1]*hpf_mask[2,2]

cv2_imshow(hpf_img)
```

### High Boost Filter

```python
hbf_mask = np.array([[-1,-1,-1],[-1,8.9,-1],[-1,-1,-1]])
hbf_mask = hbf_mask/9
print(hbf_mask)
```

```python
hbf_img = np.zeros(image.shape[:2], dtype="uint8")
for i in range(0, image.shape[0]-1):
  for j in range(0, image.shape[1]-1):
     hbf_img[i,j] = image[i-1,j-1]*hbf_mask[0,0] + image[i-1,j]*hbf_mask[0,1] + image[i-1,j+1]*hbf_mask[0,2] + image[i,j-1]*hbf_mask[1,0] + image[i,j]*hbf_mask[1,1] + image[i,j+1]*hbf_mask[1,2] + image[i+1,j-1]*hbf_mask[2,0] + image[i+1,j]*hbf_mask[2,1] + image[i+1,j+1]*hbf_mask[2,2]

cv2_imshow(hbf_img)
```
