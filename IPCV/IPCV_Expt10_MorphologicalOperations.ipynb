```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("binary.png", cv2.IMREAD_GRAYSCALE)
# cv2.imshow("Displayed Image",image)
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

```python
print("Size of image: ",image.shape)
print("Datatype of image: ",image.dtype)
print("Channels of image: ",image.ndim)
print("Total number of pixels (Size of image):",image.size)
```

```python
kernel = np.ones((5,5), np.uint8)
```

## Erosion

```python
erosion = cv2.erode(image, kernel, iterations=1)
cv2_imshow(erosion)
```

## Dilation

```python
dilation = cv2.dilate(image, kernel, iterations=1)
cv2_imshow(dilation)
```

## Opening

```python
opening = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)
cv2_imshow(opening)
```

```python
open = cv2.erode(image, kernel, iterations=1)
open = cv2.dilate(open, kernel, iterations=1)
cv2_imshow(open)
```

## Closing

```python
closing = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)
cv2_imshow(closing)
```

```python
close = cv2.dilate(image, kernel, iterations=1)
close = cv2.erode(close, kernel, iterations=1)
cv2_imshow(close)
```
