```python
finland_flag_img = cv2.imread("finland_flag.png", cv2.IMREAD_GRAYSCALE)
cv2_imshow(finland_flag_img)
```

#### Horizontal Line Detection

```python
horizontal_line = np.zeros(finland_flag_img.shape[:2], dtype="uint8")
```

```python
horizontal_mask = np.array([[-1,-1,-1],[2,2,2],[-1,-1,-1]])
print(horizontal_mask)
```

```python
for i in range(0, finland_flag_img.shape[0]-1):
  for j in range(0, finland_flag_img.shape[1]-1):
     horizontal_line[i,j] = finland_flag_img[i-1,j-1]*horizontal_mask[0,0] + finland_flag_img[i-1,j]*horizontal_mask[0,1] + finland_flag_img[i-1,j+1]*horizontal_mask[0,2] + finland_flag_img[i,j-1]*horizontal_mask[1,0] + finland_flag_img[i,j]*horizontal_mask[1,1] + finland_flag_img[i,j+1]*horizontal_mask[1,2] + finland_flag_img[i+1,j-1]*horizontal_mask[2,0] + finland_flag_img[i+1,j]*horizontal_mask[2,1] + finland_flag_img[i+1,j+1]*horizontal_mask[2,2]

cv2_imshow(horizontal_line)
```

#### Vertical Line Detection

```python
vertical_line = np.zeros(finland_flag_img.shape[:2], dtype="uint8")
```

```python
vertical_mask = np.array([[-1,2,-1],[-1,2,-1],[-1,2,-1]])
print(vertical_mask)
```

```python
for i in range(0, finland_flag_img.shape[0]-1):
  for j in range(0, finland_flag_img.shape[1]-1):
     vertical_line[i,j] = finland_flag_img[i-1,j-1]*vertical_mask[0,0] + finland_flag_img[i-1,j]*vertical_mask[0,1] + finland_flag_img[i-1,j+1]*vertical_mask[0,2] + finland_flag_img[i,j-1]*vertical_mask[1,0] + finland_flag_img[i,j]*vertical_mask[1,1] + finland_flag_img[i,j+1]*vertical_mask[1,2] + finland_flag_img[i+1,j-1]*vertical_mask[2,0] + finland_flag_img[i+1,j]*vertical_mask[2,1] + finland_flag_img[i+1,j+1]*vertical_mask[2,2]

cv2_imshow(vertical_line)
```
