```python
# Check for duplicate value
df.duplicated().any()
df[df.duplicated()]
df.duplicated().sum()
# Donâ€™t use it. Drop rows with any missing values
df.dropna()
```

```python
# Fill missing values
df['Sal'].fillna(df['Sal'].mean()) # Mean for salary
df['Experience'].fillna(df['Experience'].median()) # Median for experience
df['Per_score'].fillna(df['Per_score'].mode()[0]) # Mode for performance score
df['Age'].fillna(df['Age'].median()) # Median for age
df['Dept'].fillna(df['Dept'].mode()[0]) # Mode for department
```

```python
# Convert Joining_Date to datetime
df['Joining_Date'] = pd.to_datetime(df['Joining_Date'], errors='coerce')
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
```
