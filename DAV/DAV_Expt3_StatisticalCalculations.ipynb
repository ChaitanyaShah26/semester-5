```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# Load your dataset
df = pd.read_csv('F1DriversDataset.csv')
df.info()
```

```python
df.head()
```

```python
df.drop(['Seasons', 'Championships', 'Podiums', 'Active', 'Championship Years', 'Decade', 'Champion', 'Years_Active'], axis=1)
```

```python
# Calculate statistics
mean = df.mean(numeric_only=True)
median = df.median(numeric_only=True)
mode = df.mode(numeric_only=True).iloc[0]
std = df.std(numeric_only=True)
```

```python
print('\nMean:', mean)
print('nMedian:', median)
print('\nMode:', mode)
print('nStandard Deviation:', std)
```

```python
# Label encoding for 'company'
if 'Nationality' in df.columns:
    le = LabelEncoder()
    df ['Nationality_encoded' ] = le.fit_transform(df['Nationality' ])
    print(df[['Nationality', 'Nationality_encoded']].head(10))
```

```python
# One-hot encoding for 'Category"
pd.set_option('display.width', 1000) # Increase width
pd.set_option('display.max_columns', None) # Show all columns

if 'Nationality' in df.columns:
    category_ohe = pd.get_dummies(df['Nationality'], prefix='Nationality')
    print(category_ohe.head(10))
```

```python
statistics = df[['Pole_Rate','Start_Rate','Win_Rate','Race_Wins','Points']].agg(['mean','median'])
mode_values = df[['Pole_Rate', 'Start_Rate','Win_Rate','Race_Wins','Points']].mode().iloc[0] # mode may have multiple, take first

# Combine results
statistics.loc['mode'] = mode_values

# Show as single row (flatten)
result = statistics.transpose().reset_index()
result.columns = ['Column','Mean', 'Median', 'Mode']
print(result)
```
