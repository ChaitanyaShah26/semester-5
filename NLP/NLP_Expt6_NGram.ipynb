# NLP Experiment-6


```python
corpus = [input("Enter the corpus of statements (seperated by (eos), don't use '.'):")]
```

```python
tokens = {}
bigram = {}
for line in corpus:
    words = line.split()
    for word in words:
        if word in tokens:
            tokens[word] += 1
        else:
            tokens[word] = 1
    for i in range(len(words) - 1):
        bigram_key = (words[i], words[i+1])
        if bigram_key in bigram:
            bigram[bigram_key] += 1
        else:
            bigram[bigram_key] = 1

print("Tokens: ",tokens)
print("Bigram combinations: ",bigram)
```

```python
target_statement = input("Enter target statement (with (eos)):")
target_words = target_statement.split()
print(target_words)

bigram_prob = 1

for i in range(len(target_words) - 1):
    y = tokens[target_words[i+1]]
    x = tokens[target_words[i]]
    temp_prob = bigram[(target_words[i], target_words[i+1])] / x
    bigram_prob *= temp_prob
    print(temp_prob)

print("Bigram Probability: ",bigram_prob)
```

```python
bigram_probs = {}
for (word1, word2), count in bigram.items():
    if word1 in tokens:
        bigram_probs[(word1, word2)] = count / tokens[word1]
    else:
        bigram_probs[(word1, word2)] = 0

print("Bigram Probabilities: ", bigram_probs)
```

```python
import pandas as pd

data = []
for (word1, word2), prob in bigram_probs.items():
    data.append({'Word 1': word1, 'Word 2': word2, 'Probability': prob})

bigram_df = pd.DataFrame(data)

bigram_table = bigram_df.pivot_table(index='Word 1', columns='Word 2', values='Probability').fillna(0)

display(bigram_table)
```

```python
trigram = {}
for line in corpus:
    words = line.split()
    for i in range(len(words) - 2):
        trigram_key = (words[i], words[i+1], words[i+2])
        if trigram_key in trigram:
            trigram[trigram_key] += 1
        else:
            trigram[trigram_key] = 1

print("Trigram combinations: ", trigram)
```

```python
trigram_probs = {}
for (word1, word2, word3), count in trigram.items():
    bigram_history = (word1, word2)
    if bigram_history in bigram:
        trigram_probs[(word1, word2, word3)] = count / bigram[bigram_history]
    else:
        trigram_probs[(word1, word2, word3)] = 0

print("Trigram Probabilities: ", trigram_probs)
```

```python
target_statement = input("Enter target statement (with (eos)):")
target_words = target_statement.split()
print(target_words)

trigram_prob_target = 1

for i in range(len(target_words) - 2):
    history = (target_words[i], target_words[i+1])
    next_word = target_words[i+2]

    if history in bigram:
        if (history[0], history[1], next_word) in trigram:
            temp_prob = trigram[(history[0], history[1], next_word)] / bigram[history]
        else:
            temp_prob = 0
    else:
        temp_prob = 0
    trigram_prob_target *= temp_prob
    print(f"P({next_word} | {history[0]} {history[1]}): {temp_prob}")

print("Trigram Probability of target statement: ", trigram_prob_target)
```

```python
data_trigram = []
for (word1, word2, word3), prob in trigram_probs.items():
    data_trigram.append({'Word 1': word1, 'Word 2': word2, 'Word 3': word3, 'Probability': prob})

trigram_df = pd.DataFrame(data_trigram)
```

```python
trigram_df['History'] = trigram_df['Word 1'] + ' ' + trigram_df['Word 2']
trigram_table = trigram_df.pivot_table(index='History', columns='Word 3', values='Probability').fillna(0)
display(trigram_table)
```
