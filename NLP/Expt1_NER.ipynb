```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
text = """Google was founded by Larry Page and Sergey Brin while they were Ph.D. students at Stanford University.
It was incorporated as a privately held company on September 4, 1998, and its initial public offering (IPO) took place on August 19, 2004.
The company's mission is to organize the world's information and make it universally accessible and useful.
Google's headquarters are located at the Googleplex in Mountain View, California.
Sundar Pichai is the current CEO of Google and its parent company, Alphabet Inc."""

doc = nlp(text)
```

```python
for ent in doc.ents:
    print(ent.text,"---->",ent.label_)
```

```python
from spacy import displacy

displacy.render(doc,style="ent")
```

#### Using NLTK

```python
import nltk
from nltk import word_tokenize, pos_tag, ne_chunk
from nltk.tree import Tree

# Download required NLTK data
nltk.download('punkt_tab')

nltk.download('words')
nltk.download('maxent_ne_chunker_tab')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
# Input text
text = """
Google was founded by Larry Page and Sergey Brin while they were Ph.D. students at Stanford University.
It was incorporated as a privately held company on September 4, 1998, and its initial public offering (IPO) took place on August 19, 2004.
The company's mission is to organize the world's information and make it universally accessible and useful.
Google's headquarters are located at the Googleplex in Mountain View, California.
Sundar Pichai is the current CEO of Google and its parent company, Alphabet Inc.
"""

# Tokenize and POS-tag the text
tokens = word_tokenize(text)
pos_tags = pos_tag(tokens)

# Perform Named Entity Chunking
chunk_tree = ne_chunk(pos_tags)
```

```python
print(pos_tags)
```

| Aspect                | NLTK (Natural Language Toolkit)                                           | SpaCy                                                                      |
| --------------------- | ------------------------------------------------------------------------- | -------------------------------------------------------------------------- |
| **Purpose**           | Academic, educational, research-focused                                   | Industrial-strength, production-ready NLP                                  |
| **Ease of Use**       | More complex, with many options and manual setup                          | More user-friendly, with streamlined API                                   |
| **Performance**       | Generally slower, more flexible but less optimized                        | Faster, highly optimized for real-world use                                |
| **Pretrained Models** | Provides corpora and lexicons; requires external models                   | Comes with powerful pretrained models for tasks like tagging, parsing, NER |
| **Functionality**     | Broad range of NLP tools (tokenization, parsing, stemming, tagging, etc.) | Focused on core NLP tasks with deep learning support                       |
| **Customization**     | Highly customizable and modular                                           | Less customizable but easier to use out-of-the-box                         |
| **Community**         | Older, well-established, large community                                  | Newer but rapidly growing community                                        |
| **Common Use Cases**  | Teaching NLP concepts, prototyping, research experiments                  | Production NLP systems, text processing pipelines                          |
